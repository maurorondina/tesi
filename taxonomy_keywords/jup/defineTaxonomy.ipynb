```python
data = {}
with open('./taxonomy_to_plot.json') as f:
    data = json.load(f)

pprint(object=data, indent=3, width=170, sort_dicts=False)
```

```python
# Extract tree edges from the dict
edges = []

def get_edges(root_item, root_name=None):
    for item in root_item: # 'root_item' is a list; each item is a dict
        #if item['id-category'] != "":
        edges.append((root_name, item['name-category']+" ["+item['id-category']+"]"))  # 'id-category' , 'name-category'
        if isinstance(item, dict):
            if 'macro-categories' in item.keys():
                get_edges(item['macro-categories'], item['name-category']+" ["+item['id-category']+"]")
            elif 'categories' in item.keys():
                get_edges(item['categories'], item['name-category']+" ["+item['id-category']+"]")

get_edges(data['taxonomy'], 'TAXONOMY')

# Dump edge list in Graphviz DOT format
print('strict digraph tree {')
for row in edges:
    print('    "{0}" -> "{1}";'.format(*row))
print('}')
```
