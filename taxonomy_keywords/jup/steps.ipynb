```python
data = {}
with open('./papers_subjects.json') as f:
    data = json.load(f)

data.keys()
```

```python
subjects_id_set = set()
paper_cs = False
c = 0
c_p = 0
for paper in data['papers']:
    subjects_id_set.add((paper['primary-subject']['id_subject'], paper['primary-subject']['name_subject']))
    if paper['primary-subject']['id_subject'].startswith('cs.'):
        c_p += 1
        paper_cs = True
    for other_s in paper['other-subjects']:
        subjects_id_set.add((other_s['id_subject'], other_s['name_subject']))  
        if other_s['id_subject'].startswith('cs.'):
            paper_cs = True
    if paper_cs == False:
        print(paper)
    else:
        c+=1
print(c, c_p)
len(subjects_id_set)
```

tutti i paper appartengono ad almeno una categoria dell' A.I.
Solo 17278 hanno come 'primary-subject' una categoria dell' A.I.
